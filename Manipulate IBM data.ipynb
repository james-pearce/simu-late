# Manipulate IBM's data for optimization

```python
import pandas

ibm_data = pandas.read_csv('offers_ibm.csv')
ibm_data.head()
```

```python
ibm_data = ibm_data.drop(ibm_data.columns[0], axis=1)
```

```python
ibm_data_0 = ibm_data[['name', 'customerid', 'Product1', 'Confidence1']]
ibm_data_0 = ibm_data_0.rename(index=str, columns={'Product1': 'product', 'Confidence1': 'confidence'})
ibm_data_1 = ibm_data[['name', 'customerid', 'Product2', 'Confidence2']]
ibm_data_1 = ibm_data_1.rename(index=str, columns={'Product2': 'product', 'Confidence2': 'confidence'})
ibm_data_concat = pandas.concat([ibm_data_0, ibm_data_1])
ibm_data_concat.head()
```

```python
ibm_data_pivot = ibm_data_concat.pivot_table(index=['customerid', 'name'], columns='product', values='confidence', fill_value=0)
ibm_data_pivot.head()
```

```python
ibm_data_pivot.to_csv('offers_ibm_pivot.csv')
```
